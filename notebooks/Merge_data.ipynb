## Merge data

```python
# library imports

import numpy as np
import pandas as pd
import csv
from datetime import datetime

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
steamspy_data = pd.read_csv('/Users/sstefanovic/Documents/GitHub/LHL-Capstone_Project/data/steamspy_clean.csv')
steam_data = pd.read_csv('/Users/sstefanovic/Documents/GitHub/LHL-Capstone_Project/data/steam_data_clean.csv')

# inner merge will be used to merge SteamSpy and Steam data    
merged = steamspy_data.merge(steam_data, left_on='appid', right_on='steam_appid', suffixes=('', '_steam'))
merged.head()
```

```python
print('Merged number of rows: {}'.format(merged.shape[0]))
print('SteamSpy number of rows: {}'.format(steamspy_data.shape[0]))
print('Steam number of rows: {}'.format(steam_data.shape[0]))
```

```python
# remove overlapping columns
steam_clean = merged.drop(['name_steam', 'steam_appid'], axis=1)

# reindex to reorder columns
steam_clean = steam_clean[[
    'appid',
    'name',
    'release_date',
    'developer',
    'publisher',
    'windows',
    'mac',
    'linux',
    'required_age',
    'categories',
    'genre',
    'tags',
    'positive',
    'negative',
    'average_forever',
    'median_forever',
    'owners_lowb',
    'owners_uppb',
    'price',
    'languages_sum',
    'English',
    'German',
    'French',
    'Simplified Chinese',
    'Spanish',
    'Russian',
    'Japanese',
    'Italian',
    'Portuguese',
    'Korean',
    'Traditional Chinese',
    'Polish',
    'Turkish',
    'Dutch',
    'Czech',
    'Swedish',
    'Ukrainian',
    'Hungarian', 
    'Arabic',
    'Thai',
    'Danish',
    'Norwegian',
    'Finnish', 
    'Romanian',
    'Greek',
    'Bulgarian', 
    'Vietnamese'
]]

steam_clean = steam_clean.rename({
    'tags': 'steamspy_tags',
    'positive': 'positive_ratings',
    'negative': 'negative_ratings',
    'average_forever': 'average_playtime',
    'median_forever': 'median_playtime',
    'owners_lowb': 'owners_lower_estimation',
    'owners_uppb': 'owners_upper_estimation',
    'price': 'price_CAD'
}, axis=1)

steam_clean.head()
```

```python
steam_clean['windows'] = steam_clean['windows'].astype(bool)
steam_clean['mac'] = steam_clean['mac'].astype(bool)
steam_clean['linux'] = steam_clean['linux'].astype(bool)
```

```python
steam_clean['required_age'] = steam_clean['required_age'].fillna(0)
```

```python
steam_clean['required_age'] = steam_clean['required_age'].astype('int8')
```
