## Merge data

```python
# library imports

import numpy as np
import pandas as pd
import csv

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
steamspy_data = pd.read_csv('/Users/sstefanovic/Documents/GitHub/LHL-Capstone_Project/data/steamspy_clean.csv')
steam_data = pd.read_csv('/Users/sstefanovic/Documents/GitHub/LHL-Capstone_Project/data/steam_data_clean.csv')

# since most of the data is from SteamSpy and that is dataset with the most recent data we will use left join and we'll only keep rows with data SteamSpy    
merged = steamspy_data.merge(steam_data, left_on='appid', right_on='steam_appid', suffixes=('', '_steam'), how='left')
merged.head()
```

```python
print('Merged number of rows: {}'.format(merged.shape[0]))
print('SteamSpy number of rows: {}'.format(steamspy_data.shape[0]))
print('Steam number of rows: {}'.format(steam_data.shape[0]))
```

```python
merged = merged.drop(columns=['price', 'discount'])
```

```python
merged.head()
```

```python
# remove overlapping columns
steam_clean = merged.drop(['name_steam', 'steam_appid'], axis=1)

# reindex to reorder columns
steam_clean = steam_clean[[
    'appid',
    'name',
    'release_date',
    'developer',
    'publisher',
    'windows',
    'mac',
    'linux',
    'required_age',
    'categories',
    'genre',
    'tags',
    'positive',
    'negative',
    'average_forever',
    'median_forever',
    'owners_lowb',
    'owners_uppb',
    'is_free',
    'initialprice',
    'languages_sum',
    'English',
    'German',
    'French',
    'Simplified Chinese',
    'Spanish',
    'Russian',
    'Japanese',
    'Italian',
    'Portuguese',
    'Korean',
    'Traditional Chinese',
    'Polish',
    'Turkish',
    'Dutch',
    'Czech',
    'Swedish',
    'Ukrainian',
    'Hungarian', 
    'Arabic',
    'Thai',
    'Danish',
    'Norwegian',
    'Finnish', 
    'Romanian',
    'Greek',
    'Bulgarian', 
    'Vietnamese'
]]

steam_clean = steam_clean.rename({
    'tags': 'steamspy_tags',
    'positive': 'positive_ratings',
    'negative': 'negative_ratings',
    'average_forever': 'average_playtime',
    'median_forever': 'median_playtime',
    'owners_lowb': 'owners_lower_estimation',
    'owners_uppb': 'owners_upper_estimation',
    'is_free': 'is_game_free',
    'initialprice': 'price_USD'
}, axis=1)

steam_clean.head()
```

```python
steam_clean['price_USD'] = steam_clean['price_USD'] / 100
```
